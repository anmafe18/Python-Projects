```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
lit = pd.read_csv('./data1/LIT.csv')
```

```python
lit.head(3)
```

```python
lit['Date'] = pd.to_datetime(lit['Date'])
```

```python
lit['min_max_gain'] = lit['High'] - lit['Close']
lit
```

```python
Y = 'min_max_gain'
sns.lineplot(data=lit,x='Date',y=Y)
```

```python
sns.histplot(data=lit,x='min_max_gain')
```

```python
sns.relplot(data=lit,x='Volume',y='min_max_gain',hue='Close')
```

```python
sns.relplot(data=lit,x='Volume',y='Close',hue='min_max_gain')
```

```python
sns.lmplot(data=lit,x='Volume',y='Close')
```
