```python
SPY = pd.read_csv('./data/SPY.csv')
```

```python
SPY
```

```python
plt.plot(SPY['Close'])
```

```python
#Dif in prices each day
dif = SPY['Close'] - SPY['Close'].shift()
```

```python
plt.plot(dif)
```

```python
samp_mean = dif.mean()
samp_mean
```

```python
samp_sdev = dif.std()
samp_sdev
```

### Model Creation

```python
#252 was chosen based on the amount of rows in the SPY.csv file
mod1 = np.random.normal(samp_mean,samp_sdev,252)
mod2 = np.random.normal(samp_mean,samp_sdev,252)
mod3 = np.random.normal(samp_mean,samp_sdev,252)
mod4 = np.random.normal(samp_mean,samp_sdev,252)
mod5 = np.random.normal(samp_mean,samp_sdev,252)
```

```python
model1 = SPY['Close'][0] + mod1.cumsum()
model2 = SPY['Close'][0] + mod2.cumsum()
model3 = SPY['Close'][0] + mod3.cumsum()
model4 = SPY['Close'][0] + mod4.cumsum()
model5 = SPY['Close'][0] + mod5.cumsum()
```

```python
plt.figure(figsize = (12,8))
plt.title('STOCK REAL VALUE VS MODELINGS')
plt.xlabel('Time')
plt.ylabel('Price')

plt.plot(SPY['Close'],linewidth=3,label='Real Value')
plt.plot(model1,label='Model 1')
plt.plot(model2,label='Model 2')
plt.plot(model3,label='Model 3')
#plt.plot(model4,label='Model 4')
#plt.plot(model5,label='Model 5')

plt.legend()
plt.show()
```
