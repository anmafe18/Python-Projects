```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Se define la función de Hubble y las variantes relevantes
def Hub(z):
    H0=70
    Om=0.3
    Ol=0.7
    Or=0
    h=H0*math.sqrt(Om*(1+z)**3+Or*(1+z)**4+Ol)
    return(h)
#Se define el intervalo y paso a utilizar
z=np.linspace(0,1,101)  
#Se crea el arreglo (vacío) para H
H=[]
#Se asignan los valores correspondientes al arreglo
for i in range(len(z)):
    H.append(Hub(z[i]))
plt.plot(z,H,label="H(z)")
plt.title(r"$H=H_0\sqrt{\Omega_M(1+z)^3+\Omega_R(1+z)^4+\Omega_{\Lambda}}$")
plt.xlabel("Redshift (z)")
plt.ylabel("Función de Hubble")
plt.grid()
plt.legend()
plt.show()
```
