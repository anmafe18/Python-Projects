# Photometric Redshift

## Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

1. Given the Hubble function $$H(z)=H_0\sqrt{\Omega_{M,0}(1+z)^3+\Omega_{R,0}(1+z)^4+\Omega_{\Lambda}}.$$ 

Where z is the redshift and $H_0=70\frac{km\cdot s^{-1}}{Mpc}$.

The Omega values are: $$\Omega_{M,0}=0.3,\;\Omega_{R,0}=0,\;\Omega_{\Lambda}=0.7$$

The function will be evaluated in an interval of [0,1] in steps of $\Delta z=0.01$.

```python
#Defining the Hubble function and relevant variables
H0=70
Om=0.3
Ol=0.7
Or=0
def Hub(z):
    global H0, Om, Or, Ol 
    h=H0*math.sqrt(Om*(1+z)**3+Or*(1+z)**4+Ol)
    return(h)
#Interval and step are defined
z=np.linspace(0,1,101)  
#Creation of an empty array for H
H=[]
#Values are assigned to the array
for i in range(len(z)):
    H.append(Hub(z[i]))
plt.figure(figsize=(15,10))
plt.plot(z,H,label="H(z)")
plt.title(r"$H=H_0\sqrt{\Omega_M(1+z)^3+\Omega_R(1+z)^4+\Omega_{\Lambda}}$")
plt.xlabel("Redshift (z)")
plt.ylabel("Funci√≥n de Hubble")
plt.grid()
plt.legend()
plt.show()
```

```python
#"a" is created as a boolean variable
a=0
z_emit=[0]
for i in range(len(z)):
    a=0.01+a
    z_emit.append(a)

#Speed of light in km/s
c=300000
k=c/H0
Hi=[]
#Definition of the comoving distance function
for i in range(len(z)):
    r=(0.01*k)/(np.sqrt(Om*(1+(z_emit[i]+z_emit[i+1])/2)**3+Or*(1+(z_emit[i]+z_emit[i+1])/2)**4+Ol))
    Hi.append(r)
#Creation of arrays
X=[]
DL=[]
DA=[]
#Value assignment
for i in Hi:
    a=a+i
    X.append(a)
#Angular and Luminosity arrays are created and assigned values
for i in range(len(z)):
    dl=(1+z_emit[i])*X[i]
    da=X[i]/(1+z_emit[i])
    DL.append(dl)
    DA.append(da)
plt.figure(figsize=(15,10))
plt.plot(z,X,label="X(z)")
plt.plot(z,DL,label="DL(z)")
plt.plot(z,DA,label="DA(z)")
plt.title("$\chi,\;D_L,\;D_A$")
plt.xlabel("Redshift (z)")
plt.ylabel("Distancia")
plt.grid()
plt.legend()
plt.show()
```

Using the 4 colors, the best redshift for each galaxy is obtained

```python
#Creation of arrays for each color. Redshift between 0 and 1.
mod_gr=mod_g-mod_r
mod_gi=mod_g-mod_i
mod_gu=mod_g-mod_u
mod_gz=mod_g-mod_z
tmp_red=np.where(mod_red <= 1.)
tmp_red=tmp_red[0]
mod_gr = mod_gr[tmp_red]
mod_gi = mod_gi[tmp_red]
mod_gu = mod_gu[tmp_red]
mod_gz = mod_gz[tmp_red]
```

```python
#Previous condition for differentiating galaxies
sel=np.where(magg-magr > 0.5)
ind_sel=sel[0]
data_gr=magg[ind_sel]-magr[ind_sel]
data_gi=magg[ind_sel]-magi[ind_sel]
data_gu=magg[ind_sel]-magu[ind_sel]
data_gz=magg[ind_sel]-magz[ind_sel]
```

Errors are "created" to determine $\chi ^2$

```python
err_gr = np.sqrt(errg**2 + errr**2)
err_gr = err_gr[ind_sel]
err_gi = np.sqrt(errg**2 + erri**2)
err_gi = err_gi[ind_sel]
err_gu = np.sqrt(errg**2 + erru**2)
err_gu = err_gu[ind_sel]
err_gz = np.sqrt(errg**2 + errz**2)
err_gz = err_gz[ind_sel]
```

```python
ii=270
indi1=ind_sel[ii]
#chi squared is defined
chi = ( ((mod_gr - data_gr[indi1])**2 / err_gr[indi1]**2) +
((mod_gi - data_gi[indi1])**2 / err_gi[indi1]**2) +
((mod_gu - data_gu[indi1])**2 / err_gu[indi1]**2) +
((mod_gz - data_gz[indi1])**2 / err_gz[indi1]**2) )
#Graph in function of redshift
plt.figure(figsize=(15,10))
plt.plot(mod_red[tmp_red], chi)
plt.xlabel('Redshift',fontsize=20)
plt.ylabel('$\chi^2$',fontsize=20)
plt.show()
```

```python
photo_red=np.zeros(len(sel[0]))
#The indexes shift due to the adjustment and their minimum is selected as the photometric redshift
for ii in range(len(sel[0])):
    chi = ( ((mod_gr - data_gr[ii])**2 / err_gr[ii]**2) +
    ((mod_gi - data_gi[ii])**2 / err_gi[ii]**2) +
    ((mod_gu - data_gu[ii])**2 / err_gu[ii]**2) +
    ((mod_gz - data_gz[ii])**2 / err_gz[ii]**2) )
    best_ind=np.argmin(chi)
    photo_red[ii]=mod_red[best_ind]
line=np.linspace(0,1)
#The photometric redshift is graphed in function of the spectroscopic one, obtained from the SDSS
plt.figure(figsize=(15,10))
plt.plot(magred[sel[0]], photo_red, '+')
plt.xlabel("Redshift espectrosc√≥pico", fontsize=20)
plt.ylabel("Redshift fotom√©trico", fontsize=20)
plt.plot(line, line)
plt.xlim(0,1)
plt.show()
```
